### Original BC model (100x4)

```python
layouts = ['simple', 'unident_s']
best_bc_model_paths_orig = load_pickle("data/bc_runs_orig/best_bc_model_paths")
```

```python
cp_orig_lim_performance = CP_evaluation(best_bc_model_paths_orig, layouts, num_games=1, horizon=100)
```

```python
cp_orig_lim_performance
```

```python
for map, res in cp_orig_lim_performance.items():
    for sen, val in res.items():
        if type(val) is tuple:
            cp_orig_lim_performance[map][sen] = (val[0] * 4.0, val[1] * 4.0)
        else:
            cp_orig_lim_performance[map][sen] = val * 4.0
```

```python
cp_orig_lim_performance
```

```python
save_pickle(cp_orig_lim_performance, "data/planning/cp_evaluation_orig_lim")
```
