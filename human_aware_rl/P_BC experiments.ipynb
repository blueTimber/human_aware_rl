### Original BC model (100x4)

```python
layouts = ['simple', 'unident_s']
best_bc_model_paths_orig = load_pickle("data/bc_runs_orig/best_bc_model_paths")
```

```python
p_bc_orig_lim_evaluation = P_BC_evaluation(best_bc_model_paths_orig, layouts, num_games=1, horizon=100)
```

```python
p_bc_orig_lim_evaluation
```

```python
for map, res in p_bc_orig_lim_evaluation.items():
    for sen, val in res.items():
        if type(val) is tuple:
            p_bc_orig_lim_evaluation[map][sen] = (val[0] * 4.0, val[1] * 4.0)
        else:
            p_bc_orig_lim_evaluation[map][sen] = val * 4.0
```

```python
p_bc_orig_lim_evaluation
```

```python
save_pickle(p_bc_orig_lim_evaluation, "data/planning/p_bc_evaluation_orig_lim")
```
