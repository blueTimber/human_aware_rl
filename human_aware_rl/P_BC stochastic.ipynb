```python
for layout_name in ['simple','unident_s', 'random1']:
    
    for algo_name, algo_data in p_stochastic_bc_evaluation[layout_name].items():
        final_data[layout_name][algo_name+"_stochastic"] = algo_data

    for algo_name, algo_data in p_bc_evaluation[layout_name].items():
        final_data[layout_name][algo_name] = algo_data

    for algo_name, algo_data in bc_models_performance[layout_name].items():
        final_data[layout_name][algo_name] = algo_data
```

```python
histogram_algos_2_layout = {
    "cp": (
        ["P_BC_train+BC_test_0", "P_BC_train+BC_test_0_stochastic", "BC_train+BC_test_0", "P_BC_train+BC_test_1", "P_BC_train+BC_test_1_stochastic", "BC_train+BC_test_1"], 
        ["P_BC_test+BC_test_0", "P_BC_test+BC_test_0_stochastic", "P_BC_test+BC_test_1", "P_BC_test+BC_test_1_stochastic"]
    )
}

algos_to_name = {
    "BC_train+BC_test_0": "BC+H$_{Proxy}$",
    "BC_train+BC_test_1": None,
    "P_BC_train+BC_test_0": "P$_{BC}$+H$_{Proxy}$",
    "P_BC_train+BC_test_1": None,
    "P_BC_train+BC_test_0_stochastic": "P$_{BC}$+H$_{Proxy}^{Stoch}$",
    "P_BC_train+BC_test_1_stochastic": None,
    "P_BC_test+BC_test_0": "P$_{H_{Proxy}}$+H$_{Proxy}$",
    "P_BC_test+BC_test_1": None,
    "P_BC_test+BC_test_0_stochastic": "P$_{H_{Proxy}}$+H$_{Proxy}^{Stoch}$",
    "P_BC_test+BC_test_1_stochastic": None,
}

mean_by_algo, std_by_algo = means_and_stds_by_algo(final_data)
```

```python
from matplotlib.ticker import MultipleLocator

for hist_type, hist_algo_list in histogram_algos_2_layout.items():
    hist_algos, h_line_algo = hist_algo_list
    set_style()

    fig, ax0 = plt.subplots(1, figsize=(10,6))#figsize=(20,6))

    N = 3

    plt.rc('legend', fontsize=18.5)
    plt.rc('axes', titlesize=24)
    ax0.tick_params(axis='x', labelsize=18.5)
    ax0.tick_params(axis='y', labelsize=18.5)

    spacing = 100
    minorLocator = MultipleLocator(spacing)
    ax0.yaxis.set_minor_locator(minorLocator)
    plt.grid(which = 'both', axis="y")
        
    ind = np.arange(N)
    width = 0.1
    deltas = [-2.5, -1.5, -0.5, 1.0, 2.0, 3.0]

    for i in range(len(hist_algos)):
        delta, algo = deltas[i], hist_algos[i]
        color, hatch, alpha = get_algorithm_color(algo), get_texture(algo), get_alpha(algo)
        offset = ind + delta * width
        
        if algo == "PPO+BC_test":
            continue
        
        if algo in ["PPO_SP+PPO_SP", "PBT+PBT", "CP+CP"]:
            ax0.bar(offset, mean_by_algo[algo][:N], width, color='none', edgecolor='gray', lw=1., zorder = 0, linestyle=':')
            ax0.bar(offset, mean_by_algo[algo][:N], width, label=algos_to_name[algo], hatch="", yerr=std_by_algo[algo][:N], color='none', edgecolor='gray', linestyle=':')
        else:
            ax0.bar(offset, mean_by_algo[algo][:N], width, label=algos_to_name[algo], yerr=std_by_algo[algo][:N], color=color, hatch=hatch, alpha=alpha)

    
    ax0.hlines(final_data['simple'][h_line_algo[0]][0], xmin=-0.3, xmax=-0.2, colors="red", label=algos_to_name[h_line_algo[0]], linestyle=':')
    ax0.hlines(final_data['simple'][h_line_algo[1]][0], xmin=-0.2, xmax=-0.1, colors="green", label=algos_to_name[h_line_algo[1]], linestyle=':')
    ax0.hlines(final_data['simple'][h_line_algo[2]][0], xmin=0.05, xmax=0.15, colors="red", linestyle=':')
    ax0.hlines(final_data['simple'][h_line_algo[3]][0], xmin=0.15, xmax=0.25, colors="green", linestyle=':')

    ax0.hlines(final_data['unident_s'][h_line_algo[0]][0], xmin=0.7, xmax=0.8, colors="red", linestyle=':')
    ax0.hlines(final_data['unident_s'][h_line_algo[1]][0], xmin=0.8, xmax=0.9, colors="green", linestyle=':')
    ax0.hlines(final_data['unident_s'][h_line_algo[2]][0], xmin=1.05, xmax=1.15, colors="red", linestyle=':')
    ax0.hlines(final_data['unident_s'][h_line_algo[3]][0], xmin=1.15, xmax=1.25, colors="green", linestyle=':')

    ax0.hlines(final_data['random1'][h_line_algo[0]][0], xmin=1.7, xmax=1.8, colors="red", linestyle=':')
    ax0.hlines(final_data['random1'][h_line_algo[1]][0], xmin=1.8, xmax=1.9, colors="green", linestyle=':')
    ax0.hlines(final_data['random1'][h_line_algo[2]][0], xmin=2.05, xmax=2.15, colors="red", linestyle=':')
    ax0.hlines(final_data['random1'][h_line_algo[3]][0], xmin=2.15, xmax=2.25, colors="green", linestyle=':')

    ax0.set_ylabel('Average reward per episode')
    ax0.set_title('Performance with human proxy model')

    ax0.set_xticks(ind + width / 3)
    ax0.set_xticklabels(('Cramped Rm.', 'Asymm. Adv.', 'Coord. Ring'))
    ax0.legend(loc='best')

    handles, labels = ax0.get_legend_handles_labels()
    handles = switch_indices(0, 1, handles)
    labels = switch_indices(0, 1, labels)
    ax0.legend(handles, labels)

    # where some data has already been plotted to ax
    handles, labels = ax0.get_legend_handles_labels()

    # manually define a new patch 
    patch = Patch(facecolor='white', edgecolor='black', hatch='/', alpha=0.5, label='Switched start indices')

    # handles is a list, so append manual patch
    handles.append(patch) 

    # plot the legend
    ax0.legend(handles=handles, loc='upper right')

    ax0.set_ylim(0, 800)

    plt.savefig("data/images/planning_experiments_stochastic.png", format='png', bbox_inches='tight')
    plt.show()
```
